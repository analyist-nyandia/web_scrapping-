```python
#load  libraries
import requests
from bs4 import BeautifulSoup
```

```python
#create variable to store url
world = 'https://www.worldometers.info/geography/how-many-countries-are-there-in-the-world/'
page = requests.get(world)

print(page.status_code)
print(page.text)
```

```python
#creating a soup object to parse our received htmlresponse
soup=BeautifulSoup(page.text,'html.parser')
```

```python
#let work on html tags from our website(perform html parsing)
#getting title for our page
soup.title
```

```python
#geting all links
all_links=soup.find_all("a")
all_links
#we can use a for loop to iterate through all the link that we have
for link in all_links:
  print(link.get('href'))
  pass

```

```python
#fing the countries in the world
countries=soup.find_all('td')
countries
```

```python
# parsing p
all_p=soup.find_all('p')
all_p
```

```python
#find all tables
soup.find_all('table')
```

```python
import pandas as pd
tables = soup.find_all('table')  # Get all tables (if there are multiple)
dataframes = []
for table in tables:  # Iterate through each table
    # Extract headers
    headers = [th.text.strip() for th in table.find('thead').find_all('th')]
    # Find headers in the 'thead' section

    # Extract rows
    rows = []
    for row in table.find('tbody').find_all('tr'):  # Find rows in the 'tbody' section
        cells = row.find_all('td')
        if len(cells) > 0:
            row_data = [cell.text.strip() for cell in cells]
            rows.append(row_data)

    # Create DataFrame (only if headers and rows are present)
    if headers and rows:
        df = pd.DataFrame(rows, columns=headers)
        dataframes.append(df)
```
